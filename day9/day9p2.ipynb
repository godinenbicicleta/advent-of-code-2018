```python
import datetime
class Player():
    def __init__(self,name):
        self.name = name
        self.marbles = []
        self.score = 0 
    
    def add_score(self, marble):
        self.score += marble
    
class Marble():
    def __init__(self, value):
        self.value = value
        self.position = None
    
    def set_position(self,position):
        self.position = position
    

    
class Circle():
    def __init__ (self, numMarbles, numPlayers):
        self.positions = {}
        self.numMarbles = numMarbles
        self.numPlayers = numPlayers
        self.players = []
        self.marbles = []
    
    def initPlayers(self):
        self.players = [Player(i) for i in range(self.numPlayers)]
    
    def initMarbles(self):
        self.marbles = [Marble(i) for i in range(self.numMarbles)]

    
    def dealMarbles(self):
        for k,marb in enumerate(self.marbles):
            while k >= len(self.players):
                k -= self.numPlayers
            self.players[k].marbles.append(marb)
    
    def play(self):
        for turno in range(self.numMarbles):
            #print([m[0] for m in sorted([(i.value,
             #                             myCircle.positions[i] )for i in myCircle.positions], 
              #                          key = lambda x: x[1])])
            canica = Marble(turno)
            while turno >= len(self.players):
                turno -= self.numPlayers
            
            #print(canica.value)

            if self.positions == {}:
                self.positions[canica] = 0
                current_marble = canica
                print('turno ',turno, 'positions ', self.positions)
                continue


            if len(self.positions)<2:
                self.positions[canica] = 1
                current_marble = canica
                print('turno ',turno, 'positions ', self.positions)
                index_current = 1
                continue

            if len(self.positions)>=2:
                index_current = self.positions[current_marble]
                #print('current_index ', index_current, 'len ', len(self.positions))

                if canica.value % 23 != 0:
                    if index_current+2 == len(self.positions):
                        self.positions[canica] = index_current+2
                        current_marble = canica
                        continue
                    if index_current+2 <len(self.positions):
                        self.positions[canica] = index_current+2
                        for ca in self.positions:
                            if self.positions[ca]>= index_current+2 and ca != canica:
                                self.positions[ca]+=1
                        current_marble = canica
                        continue
                    if index_current+2 >len(self.positions):
                        posi = index_current+2
                        while posi>len(self.positions):
                            posi-=len(self.positions)
                        self.positions[canica] = posi
                        
                        for ca in self.positions:
                            if self.positions[ca]>= posi and ca != canica:
                                self.positions[ca]+=1
                        current_marble = canica
                        continue
                        
                if canica.value % 23 == 0:
                    self.players[turno].add_score(canica.value)
                    
                    if index_current-7>=0:
                        print('holi')
                    

                        for k in self.positions:
                            if self.positions[k] == index_current-7:
                                a_remover = k
                                print('cambio')
                                
                    if index_current-7<0:
                        print('noli')
                        posi = index_current-7+len(self.positions)
                        for k in self.positions:
                            if self.positions[k] == posi:
                                a_remover = k

                    self.players[turno].add_score(a_remover.value)
                    
                    if index_current-7>=0:
                        for k in self.positions:
                            if self.positions[k] == index_current-6:
                                current_marble = k
                                break

                    if index_current-7<0:
                        posi = index_current-7+len(self.positions)
                        for k in self.positions:
                            if self.positions[k] == posi:
                                current_marble = k
                                break

                    
                    ibi = self.positions[a_remover]
                    del self.positions[a_remover]
                    
                    for ca in self.positions:
                            if self.positions[ca]>= ibi:
                                self.positions[ca]+=1


        #print([m.value for m in myCircle.positions])
```
