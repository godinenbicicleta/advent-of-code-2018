```python
cells = [(i,j) for i in range(1,301) for j in range(1,301)]
serial_number = 9424
```

```python
def get_value(cell):
    rackID = cell[0]+10
    power_level = int(str(int(((cell[1]*rackID+serial_number)*rackID)/100))[-1])-5
    return power_level
    
    
```

```python
diccionario = {cell:get_value(cell) for cell in cells}
```

```python
vecinos = {cell:diccionario[cell] for cell in cells}
```

```python
for cell in cells:
    try:
        vecinos[cell]+= diccionario[(cell[0]-1, cell[1]-1)]
    except:
        pass
    try:
        vecinos[cell]+= diccionario[(cell[0]-1, cell[1]+1)]
    except:
        pass
    try:
        vecinos[cell]+= diccionario[(cell[0]-1, cell[1])]
    except:
        pass
    
    try:
        vecinos[cell]+= diccionario[(cell[0]+1, cell[1]-1)]
    except:
        pass
    
    try:
        vecinos[cell]+= diccionario[(cell[0]+1, cell[1]+1)]
    except:
        pass
    
    try:
        vecinos[cell]+= diccionario[(cell[0]+1, cell[1])]
    except:
        pass
    
    try:
        vecinos[cell]+= diccionario[(cell[0], cell[1]-1)]
    except:
        pass
    
    try:
        vecinos[cell]+= diccionario[(cell[0], cell[1]+1)]
    except:
        pass
```

```python
sumas = {}
def suma_kvecinos(cell,k):
    suma = 0
    if cell not in sumas:
        sumas[cell]={}
    if k==0:
        sumas[cell][0]=diccionario[cell]
        return diccionario[cell]

    if k in sumas[cell]:
        return sumas[cell][k]
    sumados = []    
    for i in [-k,k]:
        for j in range(-k,k+1):
            try:
                suma+= diccionario[(cell[0]+i, cell[1]+j)]
                sumados.append((cell[0]+i, cell[1]+j))
                #print('sumando ',(cell[0]+i, cell[1]+j),
                 #     'valor',diccionario[(cell[0]+i, cell[1]+j)],'suma ',suma)

            except:
                pass
    for j in [-k,k]:
        for i in range(-k,k+1):
            if (cell[0]+i, cell[1]+j) not in sumados:
                try:
                    suma+= diccionario[(cell[0]+i, cell[1]+j)]
                    #print('sumando ',(cell[0]+i, cell[1]+j),
                     #     'valor',diccionario[(cell[0]+i, cell[1]+j)],'suma ',suma)

                except:
                    pass
    sumas[cell][k] = suma+ suma_kvecinos(cell,k-1)
    return sumas[cell][k]

```

```python
resultado={}
for n,cell in enumerate(cells):
    resultado[cell]={'valor':0,'vecinos':None}
    num_vecinos = min([cell[0]-1,300-cell[0],cell[1]-1,300-cell[1]])
    for k in range(0,300):
        if k>num_vecinos:
            break
        subtotal = suma_kvecinos(cell,k)
        if subtotal > resultado[cell]['valor']:
            resultado[cell]['valor']= subtotal
            resultado[cell]['vecinos']= k
    if n%1000==0:
        print(n,end=', ')
        

            
```
