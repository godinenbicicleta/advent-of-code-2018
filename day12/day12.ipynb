```python
reglas = [{r.replace(' => ','')[2]:{'left':r.replace(' => ','')[0:2],
                                    'right':r.replace(' => ','')[3:5],
                                   'result':r.replace(' => ','')[5]} }for r in rules]
```

```python
reglas
```

```python
reglas_short = {}
for ris in reglas:
    for ri in ris:
        reglas_short[ris[ri]['left']+ri+ris[ri]['right']] = ris[ri]['left']+ris[ri]['result']+ris[ri]['right']
reglas_short
```

```python
from datetime import datetime
sumas = []
#corregir extremos

diccionario = {n:k for n,k in enumerate(initial_state)}
for num in range(-30,0):
    diccionario[num]='.'

for num in range(100,1000):
    diccionario[num]='.'
    
a = datetime.now()
minimo = -3
maximo = 100

for t in range(1000):
    
    diccionario_nuevo = {}
    
    minimos = []
    maximos = []
    for k in diccionario:
        if diccionario[k]=='#':
            minimos.append(k)
    for k in diccionario:
        if diccionario[k]=='#':
            maximos.append(k)
            
    minimo = min(minimos)
    maximo = max(maximos)
            
    for pot in range(minimo-3,maximo+2):
        

        izq=diccionario[pot-2]+diccionario[pot-1]
        
        right=diccionario[pot+1]+diccionario[pot+2]
        
        center = diccionario[pot]
        
        
        diccionario_nuevo[pot]= reglas_short[izq+center+right][2]
                    
    diccionario = {key:diccionario_nuevo[key] for key in diccionario_nuevo}
    
    
    
    minimos = []
    for k in diccionario:
        if diccionario[k]=='#':
            minimos.append(k)
    for k in diccionario:
        if diccionario[k]=='#':
            maximos.append(k)
            
    minimo = min(minimos)
    maximo = max(maximos)
    
    for num in range(minimo-5,minimo+1):
        if num not in diccionario:
            diccionario[num]='.'

    for num in range(maximo-1,maximo+6):
        if num not in diccionario:
            diccionario[num]='.'
            
    suma = 0
    for k in diccionario:
        if diccionario[k] == '#':
            suma+=k
    sumas.append((suma,t))

            
print(datetime.now()-a)
print((datetime.now()-a)*50000000000/100000)
```
