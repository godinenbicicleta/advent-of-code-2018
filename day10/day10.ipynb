```python
def get_coords(line):
    return (int(line.split('position=<')[1].split(',')[0]),
            -int(line.split(', ')[1].split('>')[0]))

def get_velocity(line):
    return (int(line.split('velocity=<')[1].split(',')[0]),
            -int(line.split('velocity=<')[1].split(',')[1].split('>')[0]))
```
