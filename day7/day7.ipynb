```python
duracion = {letra:61+ k for k,letra in enumerate('ABCDEFGHIJKLMNOPQRSTUVWXYZ')}
```

```python
class Worker():
    def __init__(self,name):
        self.name = name
        self.ocupado = False
        self.tiempo_espera = 0
        self.letra = None
    
    def start(self,letra):
        self.ocupado = True
        self.tiempo_espera += duracion[letra]
        self.letra = letra
        
    def pasa_tiempo(self):
        if self.ocupado == False:
            return None
        if self.tiempo_espera>1:
            self.tiempo_espera -= 1
            return None
        else:
            self.tiempo_espera = 0
            self.ocupado = False
            return self.letra
```
